```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv('ENB2012_data.csv')
```

```python
y1=df['Y1']
y2=df['Y2']
```

```python
x=df.drop('Y1',axis=1)
x=x.drop('Y2',axis=1)
```

```python
x.info()
y1.info()
y2.info()
```

```python
x
```

```python
# Eğitim ve test setlerine ayırma
x_train, x_test, y_train, y_test = train_test_split(x, y1, test_size=0.2, random_state=1)
```

```python
# Random Forest modelini oluşturun
rf_model = RandomForestRegressor (n_estimators=100, random_state=1)

# Modeli eğitim verileri ile eğitin
rf_model.fit(x_train, y_train)
```

```python
# Test verileri ile tahmin yapın
y_pred = rf_model.predict(x_test)
```

```python

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

# Calculate R-squared (R2)
r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2): {r2:.2f}")
```

```python
# Eğitim ve test setlerine ayırma
x_train, x_test, y_train, y_test = train_test_split(x, y2, test_size=0.2, random_state=1)
# Random Forest modelini oluşturun
rf_model = RandomForestRegressor (n_estimators=100, random_state=1)

# Modeli eğitim verileri ile eğitin
rf_model.fit(x_train, y_train)
# Test verileri ile tahmin yapın
y_pred = rf_model.predict(x_test)
```

```python
# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

# Calculate R-squared (R2)
r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2): {r2:.2f}")
```
